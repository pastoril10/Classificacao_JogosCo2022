### Criando feature vencedor

```python
#função para encontrar o time winner
def procura_winner(df):
    if(int(df['score_home']) == int(df['score_away'])):
        df['winner'] = "drow"
    elif(int(df['score_home']) > int(df['score_away'])):
        df['winner'] = "times_home"
    else:
        df['winner'] = "times_away"
    return df
```

```python
jogos = jogos.apply(procura_winner, axis='columns')
```
